```python
errors_n = 0

def error(line, message):
    
    error_message = 'ERROR @ LINE {line}: {msg}'.format(line=line, msg=message)

    global errors_n
    errors_n += 1
    
    return error_message


def errors_number():
    return errors_n

def clear_errors():
    
    global errors_n
    errors_n = 0
```

```python
Token = {'IntegerType' : 'int','FloatingType' : 'float' ,'characterType' : 'char' ,'IfSy' : 'if' ,'ElseSy' : 'else' ,
         'ReturnSy':'return','EqualitySy' : '==' , 'GreterOrEqualSy' : '>=', 'LessOrEqualSy' : '<=' , 'NotOrEqualSy' : '!=', 
         'CommentSy' : '//' , 'CommentStartSy' :'/*' ,'CommentEndSy' :'*/',
         'AssigmentSy' :'=' , 'GreaterSy':'>','LessSy':'<' ,'LCBracketSy' : '{' ,'RCBracketSY' : '}' ,
         'LBracketSy' : '(' ,'RBracketSy' : ')' ,'ColonSY' : ',','SemicolSY' : ';' , 'MinusSY' : '-' ,'PlusSy' : '+',
         'MulSy' : '*' , 'DivSy' :'/' ,} 
            #identifier ,numbers, error

def reservedWords(string ,line):  #check if the word is reserved word or identifier  
    if string == Token['IntegerType']  : return ['IntegerType',Token['IntegerType'],line]
    elif string == Token['FloatingType']  : return ['FloatingType',Token['FloatingType'],line]
    elif string == Token['characterType']  : return ['characterType',Token['characterType'],line]
    elif string == Token['IfSy']  : return ['IfSy',Token['IfSy'],line]
    elif string == Token['ElseSy']  : return ['ElseSy',Token['ElseSy'],line ]
    elif string == Token['ReturnSy']  : return ['ReturnSy',Token['ReturnSy'],line ]
    else : return ['identifier' , string,line]
def getTokens(file): #take file and return list of tokens 
    f = open(file, "r")
    f= f.read()
    line = 1
    tokenList = []
    i = 0
    while i < len(f) :

        if f[i] == '\n' : 
            line = line + 1       # count numbers 
            i = i+1
            continue

        if f[i] == '/' and f[i+1] == '/' :
            
            while f[i] != '\n'  and i < len(f)-1:
                i = i+1
            if i>=len(f)-1 :
                tokenList.append(['terminat' , 'EOF' ,line+1])
                return tokenList
            i = i-1

        elif f[i] == '/' and f[i+1] == '*' :
            i = i+2
            a = len(f)
            while f[i] != '*' and f[i+1] != '/':         #no end of of comment exception
                if f[i] == '\n' :
                    line = line + 1
                if i == a-2 :
                    return error(line, 'Unterminated comment') 
                    #break
                i = i+1
            i = i+1
                

        elif f[i] ==  '=' : 
            if f[i+1] == '=' :
                i=i+1
                tokenList.append(['EqualitySy',Token['EqualitySy'],line ] )
            else : tokenList.append( ['AssigmentSy',Token['AssigmentSy'],line ]) 

        elif f[i] ==  '>' :
            if f[i+1] == '=' :
                i=i+1
                tokenList.append( ['GreterOrEqualSy',Token['GreterOrEqualSy'],line ] )
            else : tokenList.append( ['GreaterSy',Token['GreaterSy'],line ] )

        elif f[i] ==  '<' : 
            if f[i+1] == '=' : 
                i=i+1
                tokenList.append(  ['LessOrEqualSy',Token['LessOrEqualSy'],line ])
            else : tokenList.append( ['LessSy',Token['LessSy'],line ])
                
        elif f[i] ==  '!' : 
            if f[i+1] == '=' : 
                i=i+1
                tokenList.append(  ['NotOrEqualSy',Token['NotOrEqualSy'],line ])
                
        elif  f[i] == "'" and f[i+2] == "'" :
            i = i+1 
            if f[i].isascii() :
                tokenList.append( ['character',f[i],line ])
                i =i+1
            else :
                 return error(line, number + ' syntex error')
                
        
        elif  f[i] ==  Token['LCBracketSy']  : tokenList.append( ['LCBracketSy',Token['LCBracketSy'],line ])
        elif  f[i] ==  Token['RCBracketSY']  : tokenList.append( ['RCBracketSY',Token['RCBracketSY'],line ])
        elif  f[i] ==  Token['LBracketSy']  : tokenList.append( ['LBracketSy',Token['LBracketSy'],line ])
        elif  f[i] ==  Token['RBracketSy']  : tokenList.append( ['RBracketSy',Token['RBracketSy'],line ])
        elif  f[i] ==  Token['ColonSY']  : tokenList.append( ['ColonSY',Token['ColonSY'],line ])
        elif  f[i] ==  Token['SemicolSY']  : tokenList.append( ['SemicolSY',Token['SemicolSY'],line ])
        elif  f[i] ==  Token['MinusSY']  : tokenList.append( ['MinusSY',Token['MinusSY'],line ])
        elif  f[i] ==  Token['PlusSy']  : tokenList.append( ['PlusSy',Token['PlusSy'],line ])
        elif  f[i] ==  Token['MulSy']  : tokenList.append( ['MulSy',Token['MulSy'],line ])
        elif  f[i] ==  Token['DivSy']  : tokenList.append( ['DivSy',Token['DivSy'],line ])
        

        elif f[i].isalpha() :
            word = f[i]
            i = i+1
            a = len(f)
            while i < a and f[i].isalnum() :
                word = word + f[i]
                i = i+1
            i = i-1
            tokenList.append(reservedWords(word, line))
            
        elif f[i].isdigit() :
            dot = 0
            ntype = ''
            number = f[i]
            i = i+1
            a = len(f)
            while i < a and (f[i].isdigit() or f[i]=='.')  :
                number = number + f[i]
                if f[i]== '.' :
                    dot +=1
                i = i+1
            
            if dot == 0 : ntype = 'Number'
            elif dot == 1 : ntype = 'FloatN'
            else : return error(line, number + ' syntex error')
            
            if not i < a :
                tokenList.append([ntype , number ,line])
                break
            if not f[i].isalpha()  : 
                i = i-1
                tokenList.append([ntype , number ,line])
            else :
                while f[i].isalnum() : 
                    number = number + f[i]
                    i=i+1
                i = i-1
                return error(line, number + ' cant be identifier') 
            
        elif f[i] == ' ' or f[i] == '\t' : 
            i = i+1   
            continue     #skip spaces
        else : return error(line, 'syntex error ')    
        i = i+1
    tokenList.append(['terminat' , 'EOF' ,line+1])
    return tokenList
    
```

```python
def reprresent_node(tok):
    return f'{tok}'
def reprresent_calc_node(left,op,right):
    return f'({left},{op},{right})'
def reprresent_binary_node(left,right):
    return f'({left},{right})'
def reprresent_variable_node(Type,ID,Assigment):
    return "({0} {1} {2} )".format(Type,ID,Assigment)
```

```python
#parser


#file = '5+3-2*4'

class parser :
    def __init__(self , file):
        self.tokens = getTokens(file)
        self.token = self.tokens[0]
        self.token_idx = 0
        
    def get_next_token(self):
        self.token_idx +=1
        if self.token_idx <len(self.tokens):
            self.token = self.tokens[self.token_idx]
            return self.token
            
        else: return 'end of file' 
  
    def check(self ,checker ):
        if self.token[1] == checker:
            return True 
        else : return False
        
    def accept_token(self, token):
        if self.check(token):
            self.get_next_token()
            return True
        
    def expect(self, token):
        if self.accept_token(token):
            return True
        else :
            print(error(self.token[2], 'unexpected symbol :'+ str(self.token[1])))
            return False
    

#expressions start **********************************
    def factor(self):
        if self.accept_token('('):
            ex = self.exp()
            if self.expect(')'):
                return reprresent_node(ex)
        elif self.token[0] == 'Number' :
            num = self.token[1]
            self.get_next_token()
            return reprresent_node(num)
        elif self.token[0] == 'FloatN' :
            num = self.token[1]
            self.get_next_token()
            return reprresent_node(num)
        elif self.token[0] == 'character' :
            num = self.token[1]
            self.get_next_token()
            return reprresent_node(num)
        elif self.token[0] == 'identifier' :
            num = self.token[1]
            self.get_next_token()
            return reprresent_node(num)
        else :
            print (error(self.token[2], 'Expected expression  before '+ str(self.token[1])))
            return False

    def mulOp(self):
        if self.accept_token('*'):
            return '*'
        elif self.accept_token('/'):
            return '/'
        else :
            print(error(self.token[2],'unexpected symbol :'+ str(self.token[1])))
            return False
        
    def term(self):
        left = self.factor()
        if self.check('*') or self.check('/'):
            while self.check('*') or self.check('/'):
                op = self.mulOp()
                if op :
                    right = self.factor()
                    if right:
                        left = reprresent_calc_node(left,op,right)
                    else :return False
                else : return False
            return left
        elif left:
            return left 
        else :
            #print(error(self.token[2],'unexpected symbol :'+ 'Expected int or float  after '+ str(self.token[1])))
            return False
    
    
    def addOp(self):
        if self.accept_token('+'):
            return '+'
        elif self.accept_token('-'):
            return '-'
        else : 
            print(error(self.token[2],'unexpected symbol :'+ str(token[1])))
            return False
    
    def exp(self):
        left = self.term() 
        if self.check('+') or self.check('-'):
            while self.check('+') or self.check('-'):
                op = self.addOp()
                if op :
                    right = self.term()
                    
                    if right:
                        left = reprresent_calc_node(left,op,right)
                    else :return False
                else : return False
            return left
        elif left:
            if self.token[0] in ('Number', 'FloatN' ) and self.token_idx < len(self.tokens):
                print(error(self.token[2], 'Expected operation sign insted of '+ str(self.token[1])))
                return False
            return left
        else :
            #print(error(self.token[2], 'Expected int or float  after '+ str(self.token[1]))) #delete after add variable
            return False
        
# end of expreation *********************************************************

#if_stmt : if (comparison) statement_list else_part
#else_part : ε | else if_stmt
#comparison : exp [op exp]
#comp_op : <= | < | > | >= | == | !=
    
    def comp_op(self):
         if self.accept_token('<='):
            return '<='
         elif self.accept_token('<'):
            return '<'
         elif self.accept_token('>'):
            return '>'
         elif self.accept_token('>='):
            return '>='
         elif self.accept_token('=='):
            return '=='
         elif self.accept_token('!='):
            return '!='
         else : 
            #print(error(self.token[2],'unexpected symbol :'+ str(token[1])))
            return False
        
    def comparison(self):
        left = self.exp()
        op = self.comp_op()
        if op :
            right = self.exp()
            if right :
                return reprresent_calc_node(left , op , right)
                #return True 
        return left
    def else_part(self):
        node = self.token[1]
        #print('else node'+ node)
        if node=='else' : 
           # print('inside if of else')
            #self.accept_token('else')
            self.get_next_token()
            right = self.statement()
            return reprresent_binary_node('else',right)
        else :
            #print('else' + str(self.token_idx)+ str(self.token))
           #self.token_idx = self.token_idx - 2 ######################
           # print( str(self.token_idx))
            #self.get_next_token()
            #print('elsenn' + str(self.token_idx)+ str(self.token))
            return('')
        
    def if_stmt(self):
        if self.accept_token('if'):
            if self.accept_token('('):
                compare = self.comparison()
                #print('compare'+compare)
                if self.expect(')'):
                   # print('in )')
                    nextN = self.statement()
                    #print('ne ' + nextN)
                    if nextN :
                        #self.token = self.get_next_token() 
                        last = self.else_part()
                        return reprresent_calc_node('if (' + str(compare) + ')', nextN, last)
                    else :
                        print (error(self.token[2],'Expected expression  before '+ str(self.token[1])))
                        return False
            else :
                print (error(self.token[2],' missed left pracket after if' ))
                return False
        else : return False
#end of selection statment
#return statment ********************
    #return_stmt : return   expression-stmt
    def return_stmt(self):
        if self.accept_token('return'):
            value = self.exp()
            if value:
                
                if self.accept_token(';'):
                    return reprresent_calc_node('return',value , ';')
                else :
                    print(error(self.token[2],'expected ; after '+self.tokens[self.token_idx-1][1] ))
                    return False
            else :
                print(error(self.token[2],'expected identifier or expression after return sympol'))
                return False
                    
        else :
            return False
    
#end of return statment **************
#start var_declaration******************
    #type_specifier : int | float |char
    def type_specifier(self):
        if self.accept_token('int'):
            return 'int'
        elif self.accept_token('float'):
            return 'float'
        elif self.accept_token('char'):
            return 'char'
        else :             
            return False
        
    #var_next : , ID var_next | = expression | ;
    def var_next(self):
        left = self.token[1]
        if self.accept_token(','):
            if self.token[0] == 'identifier' :
                ID = self.token[1]
                self.get_next_token()
                right =self.var_next()
                return reprresent_node("("+left + ID +str(right)+")")
            else : 
                print(error(self.token[2],"expected identifier after ',' sympol"))
                return False
        elif self.accept_token('='):
            exp =self.exp()
            if exp :
                right =self.var_next()
                if right:
                    return reprresent_binary_node(" = "+exp ,right)
                else :
                    return False
        elif self.accept_token(';'): 
            return '(;)'
        else :
            print(error(self.token[2],"expected '=' , ',' or ';' before : " +left))
            return False
         
    #var_declaration : type_specifier  ID  var_next 
    def var_declaration(self):
        Type = self.type_specifier()
        if Type :
            if self.token[0] == 'identifier' :
                ID = self.token[1]
                self.get_next_token()
                right =self.var_next()
                if right:
                        return reprresent_variable_node(Type , ID ,right)
            else :
                print(error(self.token[2],"expected identifier after type : "+Type ))
                return False
                    
        
    
    #sentence : expression  =  expression ; | expression  ;
    def sentence(self ):
        node = self.exp()
        if node :
            if self.accept_token('='):
                last = self.exp()
                if last :
                    if self.expect(';'):
                        return reprresent_calc_node(node , '=' , last + ", ; ")
            else :
                if self.expect(';'):
                    return reprresent_node(node+ ", ; ")
            
#end var_declaration******************
#start main_fun****************************

    #statement : sentence | var_declaration | scope_stmt | if_stmt | return_stmt | ;
    def statement (self):
        node = self.token[0]
        #print('node '+str(node))
        if node == 'LBracketSy' or node == 'Number' or node == 'FloatN' or node == 'character' or node == 'identifier' :
            #print('sen')
            return self.sentence()
        elif node == 'IntegerType' or node == 'FloatingType' or node == 'characterType' :
            #print('var_declaration')
            return self.var_declaration() 
        elif node =='LCBracketSy':
            #print ('scope_stmt')
            return self.scope_stmt()
        elif node =='IfSy':
            #print ('if')
            return self.if_stmt()
        elif node =='ReturnSy':
            #print('ret')
            return self.return_stmt()
        elif node =='SemicolSY':
            #print('sim')
            self.get_next_token()
            return ';'
        else : return False
     
    #statement_list : statement_list   statement | ε
    def statement_list (self):
        stat = self.statement()
        if stat :
            final = ''
            while (stat):
                final = str(final) +reprresent_node(stat)
                
                stat = self.statement()
            if stat :
                final = str(final) + str(reprresent_node(stat))
            return final
    
    #scope_stmt : { statement_list   }
    def scope_stmt (self):
        if self.accept_token('{') :
            content = self.statement_list()
            self.expect('}')
            #print('rrrr')
            #print('nn ' + str(self.token))
            return reprresent_calc_node('{',content,'}')
        #else : return False
    
    #main_fun :  int main  (  )   scope_stmt    
    def main_fun(self):
        
        if self.accept_token('int'):
            if self.expect('main') :
                if self.expect('('):
                    if self.expect(')'):
                        if self.token[1] == '{':
                            node = self.scope_stmt()
                            if node :
                                return reprresent_calc_node('int','main( )', node)
                        else :
                            print(error(self.token[2],"expected '{' after : )" ))
                            return False
        else :
            print(error(self.token[2],"main funiction not defiend" ))
            return False
            
            
            
   
```
